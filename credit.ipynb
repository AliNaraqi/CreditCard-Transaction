```python
# Simple rule-based outlier detection (e.g., amount > 99th percentile)
threshold = df['amt'].quantile(0.99)
df['is_fraud_suspect'] = df['amt'] > threshold

print(f"Flagged {df['is_fraud_suspect'].sum()} suspicious transactions (amount > 99th percentile)")
```

```python
from sklearn.ensemble import IsolationForest

# Use numeric features only
features = ['amt']
if 'lat' in df.columns and 'long' in df.columns:
    features += ['lat', 'long']

# Reset index so datetime doesn't interfere
df_model = df[features].dropna().copy()

iso = IsolationForest(contamination=0.01, random_state=42)
df_model['anomaly'] = iso.fit_predict(df_model)

# Map output: -1 = anomaly
df_model['anomaly'] = df_model['anomaly'].map({-1: 1, 1: 0})

print("Anomalies found:", df_model['anomaly'].sum())
```

```python
import numpy as np

# Fix for duplicate index labels: assign using .values and matching positions
df['is_fraud_model'] = np.nan  # Ensure the column exists

# Use the index intersection to avoid ValueError
common_idx = df.index.intersection(df_model.index)
df.loc[common_idx, 'is_fraud_model'] = df_model.loc[common_idx, 'anomaly'].values
```

```python
# Distribution of fraud vs non-fraud
sns.histplot(data=df, x='amt', hue='is_fraud_model', bins=50, kde=True, multiple='stack')
plt.title("Amount Distribution: Anomaly Detection")
plt.show()
```

```python
# View top 10 suspicious transactions
df[df['is_fraud_model'] == 1].sort_values(by='amt', ascending=False).head(10)
```

```python
# Compare average amount
df.groupby('is_fraud_model')['amt'].describe()
```

```python
df['is_fraud_model'] = df['is_fraud_model'].fillna(0)  # in case of NaNs
fraud_trend = df[df['is_fraud_model'] == 1].resample('M').size()

fraud_trend.plot(figsize=(10, 5), marker='o')
plt.title("Detected Fraudulent Transactions Over Time")
plt.ylabel("Count")
plt.xlabel("Month")
plt.grid(True)
plt.show()
```
